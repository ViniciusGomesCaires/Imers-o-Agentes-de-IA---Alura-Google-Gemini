```python
# Se a lista estiver vazia (considerado False), 'not' inverte para True
x = 10
y = 5

# Verifica se x é igual a y
igual = (x == y)
print("x é igual a y?", igual)  # Saída: False

# Inverte o valor booleano
diferente = not (x == y)
print("x é diferente de y?", diferente)  # Saída: True
```

```python
def saudar(nome):
  return f'Olá, {nome}!'

mensagem_1 = saudar('Guilherme')
mensagem_2 = saudar('Ana')
mensagem_3 = saudar('José')
```

```python
def saudar(nome: str) -> str:
  # retorna a mensagem Olá seguido do nome que recebemos como argumento
  return f'Olá, {nome}!'

mensagem_1 = saudar('Guilherme')
mensagem_2 = saudar('Ana')

print(mensagem_1)
print(mensagem_2)
```
